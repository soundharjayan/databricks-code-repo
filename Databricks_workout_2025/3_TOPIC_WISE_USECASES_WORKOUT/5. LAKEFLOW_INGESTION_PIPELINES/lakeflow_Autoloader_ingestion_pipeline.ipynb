```python
source_path = "/Volumes/soundhar_catalog/soundhar_schema/soundhar_volume/Volumes/"
schema_path1 = "/Volumes/soundhar_catalog/soundhar_schema/soundhar_volume/bronze/streamwrite13/_schema"
schema_path2 = "/Volumes/soundhar_catalog/soundhar_schema/soundhar_volume/bronze/streamwrite12/_schema"
checkpoint_path = "/Volumes/soundhar_catalog/soundhar_schema/soundhar_volume/bronze/streamwrite13/_checkpoint"
bronze_path = "/Volumes/soundhar_catalog/soundhar_schema/soundhar_volume/bronze/streamwrite13/"

df1 = (spark.readStream
       .format("cloudFiles")
       .option("cloudFiles.format","csv")
       .option("cloudFiles.maxFilesPerTrigger",1)
       .option("cloudFiles.inferColumnTypes",True)
       .option("cloudFiles.schemaEvolutionMode","addNewColumns")
       .option("cloudFiles.schemaLocation",schema_path1)
       .option("header",True)
       .load(source_path)
)

df2 = (df1.writeStream
 .trigger(availableNow=True)
 .option("checkpointLocation",checkpoint_path)
 .option("cloudFiles.schemaLocation",schema_path2)
 .option("mergeSchema","true")
 .start(bronze_path)
 )
```
