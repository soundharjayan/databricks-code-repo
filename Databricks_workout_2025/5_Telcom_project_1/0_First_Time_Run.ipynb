####Step1 (First run): Run the below Foreign Catalog Connection creation step (only for the first time)

```python
%sql
CREATE CONNECTION telecom_gcp_mysql_connection_1
TYPE mysql
OPTIONS (
  host secret('telecom_db_scope', 'db_hostname'),
  port '3306',
  user secret('telecom_db_scope', 'db_username'),
  password secret('telecom_db_scope', 'db_password')
);
```

```python
%sql
CREATE FOREIGN CATALOG IF NOT EXISTS telecom_foreign_catalog
USING CONNECTION telecom_gcp_mysql_connection_1;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS telecom_bronze;
CREATE SCHEMA IF NOT EXISTS telecom_silver;
CREATE SCHEMA IF NOT EXISTS telecom_gold;
```
