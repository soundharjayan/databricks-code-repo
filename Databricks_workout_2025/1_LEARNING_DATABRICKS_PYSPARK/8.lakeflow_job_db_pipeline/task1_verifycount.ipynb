```python
row_count=spark.sql("""select * from gcp_mysql_fc_we471.logistics.shipments1 where updated_at>(SELECT COALESCE(MAX(updated_at), '1970-01-01')  FROM catalog1_we47.schema1_we47.bronze_shipments1)""").count()
print(row_count)
```
