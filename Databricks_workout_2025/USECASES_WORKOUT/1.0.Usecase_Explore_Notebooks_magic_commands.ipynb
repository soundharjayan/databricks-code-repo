# Task 6 : Create a pyspark dataframe df1 reading the data from the above file using pyspark magic command %python 

```python
%python
df1 = spark.read.csv("dbfs:/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv",header=True)
display(df1)
```

# Task 7 : Write the above dataframe df1 data into a databricks table called 'default.mobile_os_usage' using pyspark magic command %python

```python
%python

df2 = df1.withColumnRenamed(
    "Mobile Operating System", "Mobile_Operating_System"
).withColumnRenamed(
    "Percent of Usage", "Percent_of_Usage"
)
display(df2)
df2.write.mode("overwrite").format("delta").saveAsTable("mobile_os_usage")
```
